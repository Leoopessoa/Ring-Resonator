```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from IPython.display import display, Math
```

```python
ns = 1.444
nf = 3.476
lambda0 = 1550e-9
pi = np.pi 
k0 = 2*pi/lambda0
c0 = 299792458
```

```python
figure1 = plt.figure(figsize=(12, 10))
a = 1
r = [0, 0.2, 0.4, 0.6, 0.8, 0.95]
phi = np.linspace(-pi,pi,1000)

for i in range(0,len(r)):
    phi_eff = pi + phi + np.arctan((r[i]*np.sin(phi))/(a-r[i]*np.cos(phi))) + np.arctan(a*r[i]*np.sin(phi)/(1-a*r[i]*np.cos(phi)))  
    plt.grid(True)
    plt.xlim([-pi,pi])
    plt.xticks([-pi, -pi/2, 0, pi/2, pi], ['-π', '-π/2', '0', 'π/2', 'π'], size = '15')
    plt.yticks([0, pi, 2*pi], ['0', '-π', '2π'], size = '15')
    plt.ylim([0,2*pi])
    plt.ylabel('Effective Phase Delay φ', size = '15')
    plt.xlabel('Detuning ϕ', size = '15')
    plt.plot(phi, phi_eff, label = 'r ' + str(r[i]), linewidth = 3)
    plt.legend(loc = 'upper left', prop = {'size': 15})
```

```python
figure2 = plt.figure(figsize=(12, 10))
a = [0.8, 0.85, 0.9, 0.95]
r = 0.85
phi = np.linspace(-pi,pi,1000)

for i in range(0,len(a)):
    phi_eff = pi + phi + np.arctan((r*np.sin(phi))/(a[i]-r*np.cos(phi))) + np.arctan((a[i]*r*np.sin(phi))/(1-(a[i]*r*np.cos(phi))))  
    plt.grid(True)
    plt.xlim([-pi,pi])
    plt.xticks([-pi, -pi/2, 0, pi/2, pi], ['-π', '-π/2', '0', 'π/2', 'π'], size = '15')
    plt.yticks([0, pi, 2*pi], ['0', '-π', '2π'], size = '15')
    plt.ylim([0,2*pi])
    plt.ylabel('Effective Phase Delay φ', size = '15')
    plt.xlabel('Detuning ϕ', size = '15')
    plt.plot(phi, phi_eff, label = 'a ' + str(a[i]), linewidth = 3)
    plt.legend(loc = 'upper left', prop = {'size': 15})
```
